# Lets upsample the records for "No" class as it has less count than "Yes" count to match "Yes" count

```python
import random
idx_out=[]
for i in range(0,100):
    idx = random.randint(0,100)
    idx_out.append(idx)
```

```python
Data_No = Data[Data["classification"]=='no']
```

```python
Data_No = Data_No.reset_index()
```

```python
Data_No = Data_No.drop("index",axis=1)
```

```python
Data_No_Final = Data_No.iloc[idx_out]
```

```python
Data_No_Final = Data_No_Final.reset_index()
```

```python
Data_No_Final
```

```python
Data_No_Final.drop("index",axis=1,inplace=True)
```

```python
Data_No_Final
```

```python
Data = Data.append(Data_No_Final,ignore_index=True)
```
